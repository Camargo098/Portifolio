```python
import pandas as pd


def load_data():

    base_url = 'https://raw.githubusercontent.com/Camargo098/Portifolio/main/Harry_Potter_Movies/data/'
    
    Chapters = pd.read_csv(base_url + 'Chapters.csv', sep=',', encoding='latin1')
    Characters = pd.read_csv(base_url + 'Characters.csv', sep=',', encoding='latin1')
    Data_Dictionary = pd.read_csv(base_url + 'Data_Dictionary.csv', sep=',', encoding='latin1')
    Dialogue = pd.read_csv(base_url + 'Dialogue.csv', sep=',', encoding='latin1')
    Houses = pd.read_csv(base_url + 'Houses.csv', sep=',', encoding='latin1')
    Movies = pd.read_csv(base_url + 'Movies.csv', sep=',', encoding='latin1')
    Places = pd.read_csv(base_url + 'Places.csv', sep=',', encoding='latin1')
    Spells = pd.read_csv(base_url + 'Spells.csv', sep=',', encoding='latin1')

    return Chapters, Characters, Data_Dictionary, Dialogue, Houses, Movies, Places, Spells

Chapters, Characters, Data_Dictionary, Dialogue, Houses, Movies, Places, Spells = load_data()
```

```python
character_appearances = Dialogue['Character ID'].value_counts().reset_index()
character_appearances.columns = ['Character ID', 'appearances']
df_chars = pd.merge(character_appearances, Characters, on='Character ID')

df_chars
```
