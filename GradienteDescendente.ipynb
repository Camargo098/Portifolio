<h3><b>Gradiente descendete</b></h3>
<p>O gradiente descendente é um algoritmo utilizado em aprendizado de máquina e otimização. Por meio dele, é possível reduzir a função de custo.
O gradiente descendente atua de forma iterativa para encontrar o mínimo de uma função, permitindo a convergência do modelo durante o processo de otimização.

A matémática por tráz do algoritmo é a seguinte:
Em um problema de regressão linear, estamos buscando desenvolver o modelo mais acertivo possivel. A acertividade do modelo pode ser medida atrés do <b>Erro quadratico Médio (MSE)</b>.

Vamos supor os seguintes valores:

$y$ = [2, 3, 6]

$x$ = [1, 2, 3]

$b_0, b_1 $ = 0, 0

$m$ = 3


1- Previsão com $b_0$ = 0, $b_1$ = 0:

$$\hat{y} = [0 + 0 \cdot  1, 0 + 0 \cdot 2, 0 + 0 \cdot 3]=[0,0,0] $$


Aplicando a função de custo (MSE)

$$
J(b_0, b_1) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2
$$

onde:
- $j$ representa o custo(erro)
- $m$ é o número total de exemplos(observações)
- $\sum_{i=1}^m$ é usado para obter a média da soma dos erros quadrados de todos os exemplos do conjunto.
- $\hat{y}^{(i)}$ é o valo previsto pelo modelo para o $i$-ésimo exemplo.
- $y^{(i)}$ é o valor real observado para o $i$-ésimo exemplo.

2- Erros quadraticos 
$$(0 - 2)^2 = 4, (0 - 3)^2 = 9, (0 - 6)^2 = 36 $$

3- MSE
$$ J(0,0) = \frac{4+9+26}{2 x 3} = \frac{49}{6} \approx 8.17 $$


Derivada em relação a $b_0$

$$
\frac{\partial J}{\partial b_0} = \frac{1}{m}\sum(\hat{y}^{(i)} - y^{(i)})
$$

Onde:
- $\frac{\partial J}{\partial b_0}$ representa a derivada parcial da função de custos $J$ em relação ao parâmetro $b_0$. Isso indica o quando as mudanças $b_0$ afetam o valor da função de custos.
- $\frac{1}{m}$ é o fator de normalização.


Na prática:

$$ 
\frac{(0-2)+(0-3)+(0-6)}{3} = \frac{-11}{3} \approx -3.67
$$

Derivada em relação a $b_1$
$$
\frac{\partial J}{\partial b_1} = \frac{1}{m}\sum(\hat{y}^{(i)} - y^{i}) \cdot x^{i}
$$

Na prática:

$$
\frac{(0-2) \cdot 1 + (0-3) \cdot 2 + (0-6) \cdot 3}{3} = \frac{-2 -6 -18}{3} \approx -8.67
$$

Tendo os valores das derivadas, é possivel atualizar os parametros.

Para $b_0$

$$
b_0 = b_0 - \alpha \cdot  \frac{\partial J}{\partial b_0}
$$

Onde:
- $\alpha$ representa o tamanho do passo

Na pratica:
$$
0 - 0.1 \cdot (-3.67) \approx 0.37
$$

Para $b_1$

$$
b_1 = b1 - \alpha \cdot \frac{\partial J}{\partial b_0}
$$

Na pratica:

$$
0 - 0.1 \cdot (-8.67) \approx 0.87
$$

Esse processo se repete até a convergência do modelo.

<p>É claro, existem maneiras mais facilitadas de aplicar isso na prática, variando do uso de funções que você mesmo pode escrever até o uso de funções que estão disponíveis em bibliotecas como o Scikit-learn.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_pontos = 1000

X = np.random.rand(num_pontos)

ruido = np.random.normal(0, 0.7, num_pontos)

Y = 2 * X + 3 + ruido


plt.figure(figsize=(8, 6))
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Conjunto de Dados com Correlação Positiva")
plt.grid(True)
plt.show()
```

```python
b0 = 0
b1 = 0


m = len(X)

y_previsto = b1 * X + b0
```

```python
plt.figure(figsize=(8, 6))

# Plotando a linha de tendência
plt.plot(X, y_previsto, color='red', label='Linha de Tendência (Y = 2X + 1)')
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Conjunto de Dados com Correlação Positiva")
plt.grid(True)
plt.show()

```

```python
learning_rate = 0.09
# Função de custo (MSE com 1/2)
def calcular_custo(b1, b0, X, Y):
    return np.sum((b1 * X + b0 - Y)**2) / (2 * m)

# Derivadas parciais
def calcular_derivada_b1(b1, b0, X, Y):
    return np.sum((b1 * X + b0 - Y) * X) / m

def calcular_derivada_b0(b1, b0, X, Y):
    return np.sum(b1 * X + b0 - Y) / m

# Gradiente Descendente (uma iteração)
def realizar_iteracao_gradiente_descendente(b1_atual, b0_atual, learning_rate, X, Y):
    derivada_b1 = calcular_derivada_b1(b1_atual, b0_atual, X, Y)
    derivada_b0 = calcular_derivada_b0(b1_atual, b0_atual, X, Y)
    b1_novo = b1_atual - learning_rate * derivada_b1
    b0_novo = b0_atual - learning_rate * derivada_b0
    return b1_novo, b0_novo

# Listas para armazenar o histórico do custo e parâmetros
historico_custo = []
historico_b1 = []
historico_b0 = []

# Cálculo do custo inicial
custo_inicial = calcular_custo(b1, b0, X, Y)
historico_custo.append(custo_inicial)
historico_b1.append(b1)
historico_b0.append(b0)

print(f"Custo inicial (b1={b1:.2f}, b0={b0:.2f}): {custo_inicial:.2f}")

# Loop de iterações
for i in range(50):
    b1, b0 = realizar_iteracao_gradiente_descendente(b1, b0, learning_rate, X, Y)
    custo = calcular_custo(b1, b0, X, Y)
    historico_custo.append(custo)
    historico_b1.append(b1)
    historico_b0.append(b0)
    print(f"Iteração {i+1}: b1={b1:.2f}, b0={b0:.2f}, Custo={custo:.2f}")

# Plot da curva de custo
plt.figure(figsize=(10, 6))
plt.plot(historico_custo, 'r-', linewidth=2)
plt.xlabel('Iteração', fontsize=12)
plt.ylabel('Custo (MSE/2)', fontsize=12)
plt.title('Evolução do Custo no Gradiente Descendente', fontsize=14)
plt.grid(True)
plt.show()

print(f"\nValores finais: b1 = {b1:.4f}, b0 = {b0:.4f}")
```

```python
b0_final = historico_b0[-1]
b1_final = historico_b1[-1]
y_final = b1_final * X + b0_final
```

```python
plt.figure(figsize=(8, 6))

# Plotando a linha de tendência
plt.plot(X, y_final, color='red', label='Linha de Tendência (Y = 2X + 1)')
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Conjunto de Dados com Correlação Positiva")
plt.grid(True)
plt.show()
```
